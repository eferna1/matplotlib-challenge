```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
# Study data files
mouse_metadata_path = "Pymaceuticals/data/Mouse_metadata.csv"
study_results_path = "Pymaceuticals/data/Study_results.csv"

# Read the mouse data
mouse_metadata = pd.read_csv(mouse_metadata_path)
mouse_metadata.head()
```

```python
# Read the study results
study_results = pd.read_csv(study_results_path)
study_results.head()
```

# Combine above data into single DataFrame

```python
# Merge mouse data and study results into single data frame and display a preview
mouse_data_combined = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID"])
mouse_data_combined.head()
```

```python
# Checking the number of total results
mice_count = mouse_data_combined["Mouse ID"].count()
mice_count
```

```python
# Checking the number of unique mice
unique_mice = mouse_data_combined["Mouse ID"].nunique()
unique_mice
```

# Remove Duplicate

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
dup_Mouse_ID = mouse_data_combined.loc[mouse_data_combined.duplicated(subset=["Mouse ID", "Timepoint"]), 'Mouse ID'].unique()
dup_Mouse_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_Mouse = mouse_data_combined.loc[mouse_data_combined["Mouse ID"] == "g989", :]
dup_Mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#clean_mouse_data = mouse_data_combined.drop_duplicates(subset=["Mouse ID", "Timepoint"])
clean_mouse_data = mouse_data_combined[mouse_data_combined["Mouse ID"].isin(dup_Mouse_ID)==False]
clean_mouse_data
```

```python
# Checking the number of mice in the clean DataFrame.
unique_mice = clean_mouse_data["Mouse ID"].nunique()
unique_mice
```

# Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
```

```python
# Creating a grougby for drug regimen
drug_grouped = clean_mouse_data.groupby(["Drug Regimen"])
print(drug_grouped)
```

```python
# Find the mean of the tumor volume per drug regimen
mean_tumor_vol = drug_grouped["Tumor Volume (mm3)"].mean()

# Find the median of the tumor volume per drug regimen
median_tumor_vol = drug_grouped["Tumor Volume (mm3)"].median()

# Find the variance of the tumor volume per drug regimen
var_tumor_vol = drug_grouped["Tumor Volume (mm3)"].var()

# Find the variance of the tumor volume per drug regimen
std_tumor_vol = drug_grouped["Tumor Volume (mm3)"].std()

# Find the SEM of tumor volume per drug regimen
sem_tumor_vol = drug_grouped["Tumor Volume (mm3)"].sem()
```

```python
# creating a data frame for the above - statistics summary v1
stat_sum_v1 = pd.DataFrame({"Tumor Vol Mean" : mean_tumor_vol, 
                               "Tumor Vol Median" : median_tumor_vol, 
                               "Tumor Vol Var" : var_tumor_vol, 
                               "Tumor Vol Standard Deviation" : std_tumor_vol,
                               "Tumor Vol Standard Error" : sem_tumor_vol})

# Setting drug regmin as the index
drug_regimen = clean_mouse_data.set_index(["Drug Regimen"])

# Display the first Summary Data Frame
stat_sum_v1
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
stat_sum_v2 = clean_mouse_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(['mean', 'median', 
                                                                                 'var', 'std', 'sem'])

# Setting drug regmin as the index
drug_regimen = clean_mouse_data.set_index(["Drug Regimen"])

# Display statistical summary v2
stat_sum_v2
```

# Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = clean_mouse_data.groupby(["Mouse ID"])["Timepoint"].max().reset_index()
max_timepoint = max_timepoint.rename(columns={"Timepoint" : "Max Timepoint"})

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_data = pd.merge(clean_mouse_data, max_timepoint, on="Mouse ID")
merge_data.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_list = []

# Locate the rows which contain mice on each drug and get the tumor volumes 
# add subset
for drug in treatment_list: 
    drug_data = merge_data.loc[merge_data["Drug Regimen"] == drug]
    final_tumor_volume = drug_data.loc[drug_data["Timepoint"] == drug_data["Max Timepoint"]]["Tumor Volume (mm3)"]
    tumor_volume_list.append(final_tumor_volume)
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for i, drug in enumerate(treatment_list):
    quartiles = tumor_volume_list[i].quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq    
    
    lower_bound = lowerq - 1.5 * iqr
    upper_bound = upperq + 1.5 * iqr    
   
    
    outliers = tumor_volume_list[i][(tumor_volume_list[i] < lower_bound) |
                                      (tumor_volume_list[i] > upper_bound)]
    
    print(f"{drug} Potential Outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig, ax = plt.subplots()
ax.boxplot(tumor_volume_list, labels=treatment_list, showfliers=True, flierprops=dict(marker='o', 
                        markerfacecolor='red', markersize=8))
ax.set_xlabel('Drug Regimen')
ax.set_ylabel('Final Tumor Volume (mm3)')
ax.set_title('Distribution of Final Tumor Volume by Drug Regimen')
plt.show()    
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Capomulin_data = clean_mouse_data[clean_mouse_data["Drug Regimen"] == "Capomulin"]

line_df = Capomulin_data.loc[Capomulin_data["Mouse ID"] == "m957",:]
line_df.head()
x_axis = line_df["Timepoint"]
tumor_size = line_df["Tumor Volume (mm3)"]

fig1, ax1 = plt.subplots()
plt.title("Capomulin treatmeant of mouse m957")
plt.plot(x_axis, tumor_size, linewidth=2, marker="o", color="blue")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire 
# Capomulin regimen regimen
Capomulin_data = clean_mouse_data.loc[clean_mouse_data["Drug Regimen"] == "Capomulin"]
avg_Capomulin_vol = Capomulin_data.groupby(["Mouse ID"])[["Weight (g)","Tumor Volume (mm3)"]].mean()

plt.scatter(avg_Capomulin_vol["Weight (g)"], avg_Capomulin_vol["Tumor Volume (mm3)"], color="blue")
plt.title("Mouse Weight vs Average Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

# Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(avg_Capomulin_vol["Weight (g)"], avg_Capomulin_vol["Tumor Volume (mm3)"])
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")
```

```python
# Adding the linear regression equation and line to plot
x_values = avg_Capomulin_vol["Weight (g)"]
y_values = avg_Capomulin_vol["Tumor Volume (mm3)"]
slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values,y_values, color="blue")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse Weight vs Average Tumor Volume")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
